```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
driver.get('https://www.google.com')
```

```python
driver.get('https://realpython.github.io/fake-jobs/')
```

```python
# by id 
# by class name
# by xpath 
# by javascript 
# by css selector 
# By tag name 
```

```python
title = driver.find_element(By.CLASS_NAME,'title')
```

```python
title.text
```

```python
description = driver.find_element(By.CLASS_NAME,'subtitle.is-3')
description.text
```

```python
id = driver.find_element(By.ID,'ResultsContainer').text
id
```

```python
column = driver.find_element(By.CLASS_NAME,'column.is-half')
column.text
```

```python
jobs = driver.find_elements(By.CLASS_NAME,'column.is-half')
jobs
```

```python
for job in jobs:
    print(job.text)
    break
```

```python
for job in jobs:
    job_title = job.find_element(By.CLASS_NAME,'title.is-5')
    # print(job.text)
    print("job_title:", job_title.text)
    location = job.find_element(By.CLASS_NAME,'location')
    print("location:", location.text)
    # date = job.find_element(By.CLASS_NAME,'is-small.has-text-grey')
    # print('Date and Time:' , date.text)
    #the datetime tag is retrieved by the .TAG_NAME attribute that's it
    time = job.find_element(By.TAG_NAME, "time")
    print(f"Time: {time.text}\n\n\n")
    # break
```

```python
# special find element use XPATH
driver.find_element(By.XPATH,'/html/body/section/div[2]/div/div[16]/div/div/div[2]').text
```
